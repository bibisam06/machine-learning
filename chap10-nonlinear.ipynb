```python
from sklearn.neural_network import MLPRegressor
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_percentage_error
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 입력 데이터 준비 
# X 변수 값은 1~100사이의 정수로 정한다
X = np.array(range(1,101))
print(X)
```

```python
# 출력 값 
y = 0.5*(X-50)**3-50000/X+120000
print(y)
```

```python
plt.scatter(X,y,color='b')
plt.title('nonlinear')
plt.xlabel('X')
plt.ylabel('Y')
plt.ylim(0,200000)
plt.show()
```

```python
#학습용과 테스트용 데이터 분리
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3, random_state=1234)
```

```python
print(X_train)
print(X_test)
```

```python
print(y_train)
print(y_test)
```

```python
# 피쳐 스케일링
X_train = X_train.reshape(-1,1)
X_test = X_test.reshape(-1,1)
y_train = y_train.reshape(-1,1)
y_test = y_test.reshape(-1,1)
print(X_train)
```

```python
scalerX = MinMaxScaler()
scalerX.fit(X_train)
X_train_norm = scalerX.transform(X_train)
```

```python
scalerY = MinMaxScaler()
scalerY.fit(y_train)
y_train_norm = scalerY.transform(y_train)
```

```python
x_test_norm = scalerX.transform(X_test)
y_test_norm = scalerY.transform(y_test)
```

```python
#모형화 
model = MLPRegressor(hidden_layer_sizes=(4),
                      activation='logistic',
                      solver='lbfgs',
                      max_iter=500)
```

```python
# 학습
model.fit(X_train_norm, y_train_norm)
```

```python
y_pred = model.predict(x_test_norm)
print(y_pred)
```

```python
y_pred = y_pred.reshape(-1,1)
```

```python
y_pred_inverse = scalerY.inverse_transform(y_pred)
print(y_pred_inverse)
```

```python
plt.scatter(X,y,color='g')
plt.title('nonlinear')
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(0,200000)
plt.scatter(X_test, y_pred_inverse, color='r')
plt.show()
```
