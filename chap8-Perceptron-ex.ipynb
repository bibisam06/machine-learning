```python
!pip install scikit-learn
!pip install matplotlib
from sklearn.linear_model import Perceptron 
import numpy as np
import matplotlib.pyplot as plt 
```

```python
X = np.array([[0,0],[0,1],[1,0],[1,1]])
#데이터 작성 ( Or 연산) 
Y = np.array([0,1,1,1])
```

```python
# 두 입력값 x1, x2를 x,y 축으로 표시한다면 영역이 구분되는것을 파악 가능하다
plt.scatter(X[:,0], X[:,1], c=Y)
plt.xlabel('x1')
plt.ylabel('x2')
plt.show()
```

```python
# Perceptrom 객채를 만들어서 모형화 ,데이터를 통해 학습 
# default : 학습 계수 = 1 허용오차 0.001 임 
model = Perceptron(verbose=1)
#학습 
model.fit(X,Y)
```

```python
# 결과물 출력
# 가중치 model.coef_
print(model.coef_)

# 편향 
print(model.intercept_)

#학습 수 
print(model.n_iter_)
```

```python


# 샘플 데이터와 모델 결과를 가정
# X: (n_samples, 2), Y: labels (n_samples,), model: 학습된 Perceptron

plt.scatter(X[:, 0], X[:, 1], c=Y)
plt.xlabel('x1')
plt.ylabel('x2')

w1 = model.coef_[0][0]
w2 = model.coef_[0][1]
b = model.intercept_[0]

# 결정 경계: x2 = -(w1/w2)*x1 - b/w2
x_vals = np.linspace(X[:, 0].min(), X[:, 0].max(), 100)
y_vals = -(w1 / w2) * x_vals - b / w2

plt.plot(x_vals, y_vals, c='red')
plt.title('Perceptron Decision Boundary')
plt.show()
```

```python
X_test = X
pred = model.predict(X_test)
print(pred)
```
