```python
# 6장 서포트 벡터 머신 - 파이썬 기반 분류 예제
from sklearn import svm 
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 학습용 데이터와 테스트용 데이터를 준비한다 
# 입력 데이터 X 
X_train = np.array([[2,3],[1,2]])
# 라벨링 -> 클래스 라벨링
Y_train = np.array([1,-1])
```

```python
# 테스트 데이터 
# 입력 
X_test = np.array([[3,3]])
```

```python
# 산포도 그리기 - 학습용 데이터 
plt.scatter(X_train[:,0], X_train[:,1], c=Y_train)
# 테스트용 데이터 
plt.scatter(X_test[:,0], X_test[:,1], c = 'red', marker = 'D' , s = 100) # D  : diamond
plt.xlabel('x1')
plt.ylabel('y1')
plt.xlim(0,5)
plt.ylim(0,5)
plt.show()
```

```python
# SVM 분류 모형화 -> 분류 모형은 선형분류로 설정(linear)
# SVC : Support Vector Classification, 서포트 벡터 선형분류기 
clf = svm.SVC(kernel='linear') 
```

```python
# 학습 데이터로 학습한다 
clf.fit(X_train, Y_train)
```

```python
#분류별 서포트벡터의 수 
print(clf.n_support_)
```

```python
# 서포트 벡터 출력
print(clf.support_vectors_)
```

```python
# 각 개체들의 색인 별 분류 
print(clf.classes_)
```

```python
# W : 가중치 
print(clf.coef_)
```

```python
# b : 편향
print(clf.intercept_)
```

```python
# 예측 
y_pred = clf.predict(X_test)
print(y_pred)
```
